```python
cam = cv2.VideoCapture(0)
ret, frame = cam.read()
if ret:
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame, interpolation = 'bicubic')
cam.release()
```
